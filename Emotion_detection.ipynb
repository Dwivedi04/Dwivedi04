```python
# read image 
img = cv2.imread("D:\Certificates\surprised-male-hipster-keeps-mouth-260nw-1109619839.webp")

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
height, width = img.shape[:2]
resized_img = cv2.resize(img,(3*height , 3*width), interpolation = cv2.INTER_CUBIC)

fig = plt.gcf()
fig.set_size_inches(10,17)


#plt.imshow(cv2.cvtColor(resized_img))
plt.imshow(resized_img[:,:,::-1])
plt.show()
```

```python
result = DeepFace.analyze(img , actions =['emotion', 'age', 'gender'], silent = False)
print(result)
```
